## Step 2: Load the data.

##### 4. Load the data into a list of lists

```python
#Opens the sat_scores.csv file and converts the data into a list of lists where each list contains all the data pertaining to one state

with open('../assets/sat_scores.csv') as f:
    sat_list = [x.strip().split(',') for x in f]
```

##### 9. Do any types need to be reassigned? If so, go ahead and do it.

```python
#this function changes the type of rate, verbal, and math data into an int for the sat_scores.csv file format.
def change_type(lists):
    for x in lists:
        for y in x:
            if x.index(y) == 1:
                lists[lists.index(x)][1] = int(y)
            elif x.index(y) == 2:
                lists[lists.index(x)][2] = int(y)
            elif x.index(y) == 3:
                lists[lists.index(x)][3] = int(y)
    return lists

change_type(sat_list)
       
```

##### 10. Create a dictionary for each column mapping the State to its respective value for that column.

##### 11. Create a dictionary with the values for each of the numeric columns

```python
#creates a dictionary with the label as the key and a list of scores as the value
scores_dict = {labels[x]: [y[x] for y in sat_list if y[0] != 'All'] for x in range(1,len(labels))}
```

##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
from math import sqrt
#calculates the standard deviation of an input dictionary for the key chosen
def std_dev(mydict,dickey):
    numlist = mydict.get(dickey)
    mean = sum(numlist)/len(numlist)
    sd = sqrt(sum([(n - mean) ** 2 for n in numlist])/len(numlist))
    print (dickey + " standard deviation: " + str(sd))
    return sd
    
std_dev(scores_dict,'Rate')
std_dev(scores_dict,'Verbal')
std_dev(scores_dict,'Math')
```

## Step 4: Visualize the data

##### 14. Using MatPlotLib and PyPlot, plot the distribution of the Rate using histograms.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

#creates a histogram for the rates
plt.hist(scores_dict.get('Rate'), color = 'Red', bins = range(0,101,10))
plt.title("Average Participation Rate Distribution")
plt.xlabel("Average Participation Rate")
plt.ylabel("Number of States")
plt.ylim([0,16])
plt.grid(axis='y')
```

##### 15. Plot the Math distribution

```python
#creates a histogram for the math scores
plt.hist(scores_dict.get('Math'), color = 'Blue', bins = range(400,650,25))
plt.title("Average Math Score Distribution")
plt.xlabel("Average Math Score")
plt.ylabel("Number of States")
plt.ylim([0,16])
plt.grid(axis='y')

```

##### 16. Plot the Verbal distribution

```python
#creates a histogram for the verbal scores

plt.hist(scores_dict.get('Verbal'), color = 'Green', bins = range(400,650,25))
plt.title("Average Verbal Score Distribution")
plt.xlabel("Average Verbal Score")
plt.ylabel("Number of States")
plt.ylim([0,14])
plt.grid(axis='y')

```

##### 19. Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
# creates a scatter plot, plotting math and verbal scores, with the size of the dot determined by the participation rate
x = scores_dict.get('Math')
y = scores_dict.get('Verbal')
s = scores_dict.get('Rate')
plt.scatter(x,y,s, marker = 'o', color='orange',label = 'Rate')
plt.title("Average Math and Verbal Score Scatter Plot", y=1.02)
plt.xlabel("Average Math Score")
plt.ylabel("Average Verbal Score")
plt.ylim([450,600])
plt.xlim([425,625])
plt.grid()
plt.legend(loc = 'upper left', fontsize = 'small')
```

##### 20. Are there any interesting relationships to note?

##### 21. Create box plots for each variable. 

```python
matharray = np.array(scores_dict.get('Math'), dtype = np.int64)
verbalarray = np.array(scores_dict.get('Verbal'), dtype = np.int64)
ratearray = np.array(scores_dict.get('Rate'), dtype = np.int64)

fig = plt.figure(figsize=(10,5))
fig.suptitle('Box Charts for Mean SAT Scores and Rate by State for 2001', fontsize = 'large',y=1.05)

ax1 = fig.add_subplot(1,3,1)

ax1.boxplot(matharray,showmeans=True)
plt.title('Math Score')


ax2 = fig.add_subplot(1,3,2)
ax2.boxplot(verbalarray,showmeans=True)
plt.title('Verbal Score')

ax3 = fig.add_subplot(1,3,3)
ax3.boxplot(ratearray,showmeans=True)
plt.title('Rate')
```
