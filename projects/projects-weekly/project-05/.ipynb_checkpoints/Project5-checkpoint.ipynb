```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
loans = pd.read_csv('assets/LoanStats3d.csv',header=1)
```

```python
def gb (x):
    if x == 'Fully Paid':
        return 'Good'
    elif x == 'Default' or x == 'Charged Off':
        return 'Bad'
    else:
        return np.nan

loans['good_bad'] = loans['loan_status'].apply(gb)
loans['good_bad'].unique()
```

```python
loans2 = loans.loc[:421094, \
            ['good_bad', 'int_rate','term','annual_inc','dti','funded_amnt','home_ownership']]

print loans['good_bad'].unique()
print loans['int_rate'].unique()
print loans['term'].unique()
print loans['dti'].unique()
print loans['funded_amnt'].unique()
print loans['home_ownership'].unique()
```

```python
def nop (x):
    if '%' in str(x):
        return str(x).strip('%').replace(" ","")
    else:
        return x
loans2["int_rate"] = loans2["int_rate"].apply(nop).astype(float).apply(lambda x: x/100)
print (loans2['int_rate'][0])
```

```python
def ft (x):
    return str(x).strip(" ").split(" months")[0]

loans2['term'] = loans2['term'].apply(ft).astype(float)
loans2['term'].unique()
```

```python
loans2['annual_inc'].astype(float)
loans2['dti'].astype(float)
loans2['funded_amnt'].astype(float)
loans2.dtypes
```

```python
def mil (x):
    if x > 300000:
        return 300000
    else:
        return x

loans2['annual_inc'] = loans2['annual_inc'].apply(mil)

```

```python
loans2['counter'] = 1
```

```python
fig,ax = plt.subplots(4,2, figsize=(16,24))
loans2[(loans2['good_bad']=='Good')&(loans2['term']==36)&(loans2['home_ownership']=='OWN')].plot\
        (x='int_rate',y = 'annual_inc', kind='scatter',s=loans2['dti'], color='blue',ax=ax[0][0])
loans2[(loans2['good_bad']=='Bad')&(loans2['term']==36)&(loans2['home_ownership']=='OWN')].plot(x='int_rate',y = 'annual_inc', kind='scatter',\
        s=loans2['dti'], color='red',ax=ax[0][0])


```
