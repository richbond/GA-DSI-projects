```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import emcee
import corner
%matplotlib inline
```

```python
# Import the data into a pd DataFrame called loans.  Since the first row is blank, I set the header to 1.
loans = pd.read_csv('assets/LoanStats3d.csv',header=1)
```

```python
# Created a function that classifies Good and Bad Loans, then added a column to loans df /
# indicating good and bad loans.  printed the unique items in the new good_bad category \
# to confirm the function worked properly.
def gb (x):
    if x == 'Fully Paid':
        return 'Good'
    elif x == 'Default' or x == 'Charged Off':
        return 'Bad'
    else:
        return np.nan

loans['good_bad'] = loans['loan_status'].apply(gb)
print (loans['good_bad'].unique())
```

```python
# Created a new df called loans2 that eliminates the total rows at the bottom and only \
# contains the columns of data I used to build and test my model.
loans2 = loans.loc[:421094, \
            ['good_bad','term','home_ownership','int_rate','annual_inc','dti','funded_amnt']]
```

```python
# Created a function that converts the interest rate to a float.  Then I applied the function \
# to the int_rate column and printed the first row to make sure it worked.
def nop (x):
    if '%' in str(x):
        return str(x).strip('%').replace(" ","")
    else:
        return x
loans2["int_rate"] = loans2["int_rate"].apply(nop).astype(float).apply(lambda x: x/100)
print (loans2['int_rate'][0])
```

```python
# coverted term strings to floats, applied it to the term column, then printed the unique \
# values to make sure it worked.

def term10(x):
    if '36' in x:
        return 36
    elif '60' in x:
        return 60
    else:
        return np.nan
    
loans2['term'] = loans2['term'].apply(term10).astype(float)
print (loans2['term'].unique())
```

```python
# Normalized the necessary terms using min/max
nc = [x for x in loans2.columns if x != 'good_bad' and x != 'term' and x != 'home_ownership']
loans2.ix[:, nc] = (loans2.ix[:, nc] - loans2.ix[:, nc].min())\
                   / (loans2.ix[:, nc].max() - loans2.ix[:, nc].min())
```

```python
# Created Dummy Variable for appropriate columns
for x in loans2.columns:
    if x != 'int_rate' and x != 'annual_inc' and x!= 'funded_amnt' and x!= 'dti':
        pd.get_dummies(loans2[x])
        loans2 = pd.concat([loans2, pd.get_dummies(loans2[x])], axis=1)
```

```python
# Created a counter column that contains the value 1 for every row.  I use this when I make a\
# pivot table and set the values to counter.  Then print the head to make sure the normalizing \
# and dummy and counter worked properly
loans2['counter'] = 1
loans2.head()
```

```python
# Created a new table of only all the closed loans defined by Good and Bad.
# Reset and drop the index.  Then print the head to confirm results
loansgb = loans2[(loans2['good_bad'] == 'Good') | (loans2['good_bad'] == 'Bad')]
loansgb.reset_index(inplace=True)
loansgb = loansgb.drop('index',axis=1)
loansgb.head()
```

```python
# Create a pivot table to explore data
# from this pivot table, I can see that 36 month loans have much lower default rates.
# also, renters are more likely to default than home owners/mortgages.
# this pivot table show me that 76% of all loans turned out to be good loans.
# This overweighting of Good loans will make it harder to predict bad loans.
loansgb.pivot_table(index=['good_bad'],values=['counter'], \
                   columns=['home_ownership','term'],\
                   aggfunc = [sum], margins=True) 
```

```python
# This process, in the end, creates a dataframe of an equal number of good and bad loans.
# Since there are much less bad loans than good, I used all of the bad loans and took the\
# same amount of good loans, then put them together in one df called loans5050.
# I resetted the index along the way.

goodloans = loansgb[loansgb['Good']==1]
badloans = loansgb[loansgb['Bad']==1]

badloans.reset_index(inplace=True)
badloans = badloans.drop('index',axis=1)

goodloans.reset_index(inplace=True)
goodloans = goodloans.drop('index',axis=1)

goodeqbad = goodloans.iloc[:24319,:]

tocombine = [goodeqbad,badloans]
loans5050 = pd.concat(tocombine)
loans5050.reset_index(inplace=True)
loans5050 = loans5050.drop('index',axis=1)
```

```python
# Sigmoid Function to put probabilities on 0-1 scale
sig = lambda x: 1./(1+np.exp(-x))
```

```python
# Create List of Columns to use for testing.  Not all columns will be used.\
# Print list of column names
xc = list(loans5050.iloc[:,3:-1].columns)
print (xc)
```

```python
# Create a list of lists where each sublist contains a row of data using the column \
# names in xc
xlist = []
for x in loans5050[xc].values:
    xlist.append(list(x))
```

```python
# Create my predictive distribution function.

# Commenting which index of the list refers to which data point.  Not using \
# all of the data such as both of the terms
# data [0] = int rate, data [1] = ann inc, data [2] = dti,  data [3] = fundedamt,\
# data [4] = bad,      data [5] = good,    data [6] = 36,   data [7] = 60,\
# data [8] = any,      data [9] = mort,    data [10] = own, data [11] = rent \

def lnpred(w):
    for e in xlist:
        P = w[0] + w[1]*e[0] + w[2]*e[1] + w[3]*e[2] + w[4]*e[3] + w[5]*e[6] + \
            w[6]*e[7] + w[7]*e[10]
        if e[5] == 1:
            return np.log(sig(P))
        else:
            return np.log(1-sig(P))
```

```python
# Create my ln prob function
def lnprob(w):
    A = np.sum([e**2 for e in w])
    return -0.5*A + np.sum(lnpred(w))
```

```python
# Sampler function and Monte Carlo
p0 = np.random.rand(20*8).reshape((20,8))
sampler = emcee.EnsembleSampler(20, 8, lnprob)
pos, prob, state = sampler.run_mcmc(p0, 1000)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 20000)
samples = sampler.flatchain
MonteCarlo = lambda f,s: 1/float(len(samples))*sum([f(e) for e in samples])
```

```python
# Calculates optimal omega values for each model parameter and prints results
Omega0 = MonteCarlo(lambda x: x[0],samples)
Omega1 = MonteCarlo(lambda x: x[1],samples)
Omega2 = MonteCarlo(lambda x: x[2],samples)
Omega3 = MonteCarlo(lambda x: x[3],samples)
Omega4 = MonteCarlo(lambda x: x[4],samples)
Omega5 = MonteCarlo(lambda x: x[5],samples)
Omega6 = MonteCarlo(lambda x: x[6],samples)
Omega7 = MonteCarlo(lambda x: x[7],samples)

print ('Omega0: {0}, Omega1: {1}, Omega2: {2}, Omega3: {3}, Omega4: {4}, Omega5: {5},\
         Omega6: {6}, Omega7: {7}').format(Omega0, Omega1,Omega2,Omega3,Omega4,\
        Omega5,Omega6,Omega7)
```

```python
# Plot each of the model parameter distributions
params = pd.DataFrame(samples, columns = ['Omega0', 'Omega1','Omega2','Omega3','Omega4',\
        'Omega5','Omega6','Omega7'])
params.plot(kind = 'density',figsize = (8,4)).set_title('Model Parameter Distributions',y=1.02)
plt.show()
```

```python
# Predict the probability of one loan being good.  Enter the list index below
predict = lambda d: MonteCarlo(lambda w: lnpred(d),samples)
np.exp(predict(xlist[47668])) # Enter the list index here to predict.
```

```python
# Calcualtes the average probability of the loan being good for the first 500 rows. \
# First 500 rows is good loan data.
# This tells me if my model is good or not
N = len(xlist[:500])
sum([np.exp(predict(x)) for x in xlist[:500]])/N
```

```python
# Calcualtes the average probability of the loan being bad for the last 500 rows. \
# Last 500 rows is bad loan data.
# This tells me if my model is good or not
N = len(xlist[-500:])
sum([np.exp(predict(x)) for x in xlist[-500:]])/N
```

Unfortunately, my model can predict good loans pretty well, however it tells me that my bad loans have a high probability of being repaid.
