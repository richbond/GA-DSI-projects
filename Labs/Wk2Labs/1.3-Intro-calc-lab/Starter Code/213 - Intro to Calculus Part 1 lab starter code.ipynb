Plot the following functions in a 4x4 block using lambda expressions, pyplot.subplots and figure.subplots_adjust (only use the matplotlib libary). Determine the appropriate plotting range and determine the image of the Function. Label each graph appropriatly including ledgend, x-labels, y-labels and title. Which one of the functions is a one-to-one map?
1. $F_1(x) = Exp(-0.5*(x-2)^2)$
2. $F_2(x) = x^2 * Exp(-0.8*(x+2)^2)$
3. $F_3(x) = F_2(x)*F_1(x)$
4. $F_4(x) = \frac{1}{1+Exp(-x)}$

Consider the following mapping:
FN:ℝ→ℝ;FN(x)=∑k=0N1k!∗xk
FN:R→R;FN(x)=∑k=0N1k!∗xk
 
This function actually represents a class of functions for any integer  N∈ℕN∈N . Let's try to find out what the image  FN(ℝ)FN(R)  looks like. It is clear that if we choose a positive  xx  value then  FN(x)>0FN(x)>0 .

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

X = np.arange(-10,11,.1)
F1_of_X = lambda x: np.exp(-.5 * ((x-2)**2))
F2_of_X = lambda x: (x**2)*(np.exp(-.8*((x+2)**2)))
F3_of_X = lambda x: F2_of_X(x) * F1_of_X(x)
F4_of_X = lambda x: 1 / (1 + np.exp(-x))

fig, axes = plt.subplots(2,2, figsize=(6,6))
fig.subplots_adjust(wspace = 0.3, hspace = 0.5)
axes[0,0].plot(X, F1_of_X(X), label = 'F1')
axes[0,1].plot(X, F2_of_X(X), label = 'F2')
axes[1,0].plot(X, F3_of_X(X), label = 'F3')
axes[1,1].plot(X, F4_of_X(X), label = 'F4')
i = 1
for e in axes:
    for ee in e: # this second loop is because it is a 2 row graph array
        ee.legend(loc = 'best', fontsize = 'small')
        ee.set_xlabel('x')
        ee.set_title('Function: F{0}'.format(i))
        i = i+1
    



```
