```python
# Create a histogram by creating a list comprehension + a function to turn the data into inches

htlist = [x for x in Height if type(x) != float]
htlist


                
```

```python
def cleanfunc(somelist):
    cleanlist = []
    for x in somelist:
        if x == '6\'6" and above':
            x = '6\'6"'
        elif x == 'Under 5 ft.':
            x = '5\'0"'
        x = x.replace('"',"")
        x = x.split("'")
        cleanlist.append(x)
    return cleanlist

cleanlist = cleanfunc(htlist)

inches = [(12*int(x[0])) + int(x[1]) for x in cleanlist]
inches
plt.hist(inches)
```
