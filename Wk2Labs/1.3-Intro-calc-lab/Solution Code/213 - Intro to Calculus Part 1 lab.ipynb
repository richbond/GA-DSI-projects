*** Exercise 1: ***
Plot the following functions in a 4x4 block using lambda expressions, pyplot.subplots and figure.subplots_adjust (only use the matplotlib libary). Determine the appropriate plotting range and determine the image of the Function. Label each graph appropriatly including ledgend, x-labels, y-labels and title. Which one of the functions is a one-to-one map?
1. $F_1(x) = Exp(-0.5*(x-2)^2)$
2. $F_2(x) = x^2 * Exp(-0.8*(x+2)^2)$
3. $F_3(x) = F_2(x)*F_1(x)$
4. $F_4(x) = \frac{1}{1+Exp(-x)}$

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

F1 = lambda x: np.exp(-0.5*(x-2)**2)
F2 = lambda x: x**2 * np.exp(-0.8* (x+2)**2)
F3 = lambda x: F1(x)*F2(x)
F4 = lambda x: 1/(1+np.exp(-x))
x = np.arange(-10,10,0.1)
fig, axes = plt.subplots(2,2)
fig.subplots_adjust(wspace = 0.3, hspace = 0.5)
axes[0,0].plot(x, F1(x), label = 'F1')
axes[0,1].plot(x, F2(x), label = 'F2')
axes[1,0].plot(x, F3(x), label = 'F3')
axes[1,1].plot(x,F4(x), label = 'F4')
i = 1
for e in axes:
    for ee in e:
        ee.legend(loc = 'best', fontsize = 'small')
        ee.set_xlabel('x')
        ee.set_title('Function: F{0}'.format(i))
        i = i+1
```

*** Exercise 2: ***
Plot the following functions and their respective contour plots in a 4x4 block (only use the matplotlib libary). Determine the appropriate plotting range and determine the image of the Function. Label each graph appropriatly including ledgend, x-labels, y-labels and title.
1. $G_1(x,y) = Exp(-0.01*(x^3+y^3))$
2. $G_2(x,y) = Log(x^2+2*y^2)*x^2$


```python
G1 = lambda x, y: np.exp(-0.01*(x**3 + y**3))
G2 = lambda x, y: np.log(x**2 +2*y**2)*x**2
from mpl_toolkits.mplot3d import Axes3D
X = np.linspace(-2,2, 100)
Y = np.linspace(-2,2, 100)
X, Y = np.meshgrid(X,Y)
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1, projection = '3d')
ax1.plot_surface(X, Y, G1(X,Y))
ax1.set_title('G1 - 3D Plot')
ax2 = fig.add_subplot(2,2,2)
ax2.contour(X, Y, G1(X,Y))
ax2.set_title('G1 - Contour Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax3 = fig.add_subplot(2,2,3, projection = '3d')
ax3.plot_surface(X, Y, G2(X,Y))
ax3.set_title('G2 - 3D Plot')
ax4 = fig.add_subplot(2,2,4)
ax4.contour(X, Y, G2(X,Y))
ax4.set_title('G2 - Contour Plot')
ax4.set_xlabel('x')
ax4.set_ylabel('y')
```
